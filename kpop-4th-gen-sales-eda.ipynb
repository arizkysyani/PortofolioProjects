```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/kpop-4th-gen-sales/Kpop 4th gen Sales - Sheet1.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
### changing types(sales) from object into int

df['date'] = pd.to_datetime(df['date'])
df['sales'] = df['sales'].str.replace(',', '').str.strip().astype('int64')
df.info()
```

```python
### checking null values

df.isnull().sum().sort_values(ascending=False)
```

```python
df.describe(include='object').T
```

checking basic statistics

```python
### Create a new column for month, day, and year

from datetime import datetime as dt

df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day_name()
df['year'] = df['date'].dt.year

df.head()
```

```python
print(df.corr())
sns.heatmap(df.corr())
```

```python
sns.lineplot(df,x='year', y='sales')
```

Average sales kpop 4th gen keep increasing from 2017 - 2021

```python
print(df.groupby(['month'])['sales'].count().sort_values(ascending=False))

plt.figure(figsize=(11,5))
sns.barplot(
    df.groupby(['month'])['sales'].count().sort_values(ascending=False).reset_index(),
    x='month', y='sales')
```

Throughout 2017 - 2021, March, October, and January are the most high selling songs

```python
print(df.pivot_table(index='Artist',
                     columns='day', 
                     values='sales', 
                     aggfunc='count', 
                     fill_value='-'))

plt.figure(figsize=(11,5))
sns.barplot(
    df.groupby(['day'])['sales'].count().sort_values(ascending=False).reset_index(),
    x='day', y='sales')
```

Most of the time Artist release their songs on Monday

```python
print(df.groupby(['Artist'])['sales'].sum().reset_index())

plt.figure(figsize=(10,5))
sns.barplot(
    df.groupby(['Artist'])['sales'].sum().reset_index(),
    x='Artist',
    y='sales')
```

IZONE is the most selling Artist from 2017 - 2021

```python
print("Artis's sales through out the year")
print(df.pivot_table(index ='Artist',
                       columns='year',
                       values ='sales',
                       aggfunc ='sum',fill_value='-'))
print('='*70)
print('How many times Artists release their songs for the past 5 years')
print(df.pivot_table(index ='Artist',
                       columns='year',
                       values ='sales',
                       aggfunc ='count',fill_value='-'))

```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    df.groupby(['Artist', 'year'])['sales'].count().reset_index(), 
    x='Artist', 
    y='sales', 
    hue='year').set(title='Sales Artist by Year')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    df.groupby(['Artist', 'year'])['sales'].sum().reset_index(), 
    x='Artist', 
    y='sales', 
    hue='year').set(title='Sales Artist by Year')
```

Except for IZONE, the total of songs they release and sales are quite competitive

**Insight:**

*  Average sales kpop 4th gen keep increasing from 2017 - 2021
* Throughout 2017 - 2021, March, October, and January are the most high selling songs, and they released it mostly on Monday
* IZONE is the most selling Artist from 2017 - 2021, and their peak sales is on 2020 with a very big gap with other artists
* Except for IZONE, the total of songs they release and sales are quite competitive from each other
* Theres no strong correlation between how many times artist release their songs and the total sales
